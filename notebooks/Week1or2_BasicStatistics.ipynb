## Grouping Data By Categories
Grouping is another one of the basic data investigation tools. We will use the groupby operator. It splits a dataset into the categories we specify, then applies whatever functions we tell it to apply, and then combines the results for each group in the output.

```python
# Using groupby to aggregate the data by one of the attributes, here sex.
# All this produces is another object, not something that's really useful to us.
adult.groupby(['sex'])
```

```python
# Let's use groupby with an aggregating function
adult.groupby(['sex']).sum()
```

```python
# Let's use groupby with another aggregating function because WHOA! Interesting results!
# Let's add more columns to the array!
adult.groupby(['educationyears','sex']).median()
```

```python
# Make the outliers pretty
green_diamond = dict(markerfacecolor='g', marker='D')
fig2, ax2 = plt.subplots()
ax2.set_title('Changed Outlier Symbols')
ax2.boxplot(adult.age, flierprops=green_diamond)
```

```python
#Flip the plot
red_square = dict(markerfacecolor='r', marker='s')
fig3, ax3 = plt.subplots()
ax3.set_title('Horizontal Boxes')
ax3.boxplot(adult.age, vert=False, flierprops=red_square)
```

## Distance in Python

EUCLIDIAN DISTANCE is the â€œordinaryâ€ straight-line distance between two points.

if p = (p1, p2) and q = (q1, q2) then the distance is given by
d(p,q) = sqrt((q1-p1)**2 + (q2-p2)**2)

For three dimensions, that is
d(p,q) = sqrt((q1-p1)**2 + (q2-p2)**2 + (q3-p3)**2)

This is what it looks like in Python

MANHATTAN DISTANCE is the distance between two points measured along axes at right angles.

distance += abs(x_value - x_goal) + abs(y_value - y_goal)

where x_value, y_value is where you are and x_goal, y_goal is where you want to go.

Here is an example:

Input : n = 4
        point1 = { -1, 5 }
        point2 = { 1, 6 }
        point3 = { 3, 5 }
        point4 = { 2, 3 }
Output : 22
Distance of { 1, 6 }, { 3, 5 }, { 2, 3 } from 
{ -1, 5 } are 3, 4, 5 respectively.
Therefore, sum = 3 + 4 + 5 = 12

Distance of { 3, 5 }, { 2, 3 } from { 1, 6 } 
are 3, 4 respectively.
Therefore, sum = 12 + 3 + 4 = 19

Distance of { 2, 3 } from { 3, 5 } is 3.
Therefore, sum = 19 + 3 = 22.
